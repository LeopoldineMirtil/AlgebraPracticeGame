# Algebra Practice Game

```python
import random
import numpy as np

score = 0

def add():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    player = int(input(f'{a} + {b} ='))
    answer = a + b
    if player == answer:
        score += 1
        print(f'Correct! The sum is {answer}') 
    else:
        print(f'Wrong, the sum is {answer}')

def sub():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    player = int(input(f'{b} - {a} ='))
    answer = b - a
    if player == answer:
        score += 1
        print(f'Correct! The difference is {answer}') 
    else:
        print(f'Wrong, the difference is {answer}')

def mult():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    player = int(input(f'{a} x {b} ='))
    answer = a*b
    if player == answer:
        score += 1
        print(f'Correct! The product is {answer}') 
    else:
        print(f'Wrong, the product is {answer}') 

def div(): 
    a = random.randint(1, 10) #prevent nums from being 0 -->error
    b = random.randint(-10, 10)
    print("Round answer to 2 decimal places when necessary")
    player = float(input(f'{b}/{a} = '))
    answer = np.around(b/a, decimals=2)
    if player == answer:
        score += 1
        print(f'Correct! The quotient is {answer}') 
    else:
        print(f'Wrong, the quotient is {answer}')

def add_sub():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(-10, 10)
    player = int(input(f'{a} + {b} - {c} ='))
    answer = a+b-c
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')

def add_mult():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(-10, 10)
    player = int(input(f'{a} + {b} x {c} ='))
    answer = a+b*c
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')

def add_div():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(1, 10)
    print("Round answer to 2 decimal places when necessary")
    player = float(input(f'{a} + {b}/{c} ='))
    answer = np.around(a+b/c, decimals=2)
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')

def sub_mult():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(-10, 10)
    player = int(input(f'{a} - {b} x {c} ='))
    answer = a-b*c
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')

def sub_div():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(1, 10)
    print("Round answer to 2 decimal places when necessary")
    player = float(input(f'{a} - {b}/{c} ='))
    answer = np.around(a-b/c, decimals=2)
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')

def mult_div():
    a = random.randint(-10, 10)
    b = random.randint(-10, 10)
    c = random.randint(1, 10)
    print("Round answer to 2 decimal places when necessary")
    player = float(input(f'{a} x {b}/{c} ='))
    answer = np.around(a*b/c, decimals=2)
    if player == answer:
        score += 1
        print(f'Correct! The answer is {answer}') 
    else:
        print(f'Wrong, the answer is {answer}')


#call functions in random order
def random_order(*m):
    select_math = list(m)
    random.shuffle(select_math)
    for m in select_math:
        m()
random_order(add, sub, mult, div, add_sub, add_mult, add_div, sub_mult, sub_div, mult_div)

print("***Game Over***")
print(f'You got {score} out of 10')
    
```
